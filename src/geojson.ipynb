```python
import numpy as np
import geopandas as gpd
# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as patches
import matplotlib
from distinctipy import distinctipy
import geopandas as gpd
import os
import numpy as np
from IPython.display import JSON
import json
import pyproj

from utils.coco_mask_utils import *

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
regions = ["aa", "bb", "Co", "dd",
            "ee", "ff", "gg", "hh",
            "ii", "jj", "kk", "A",
            "B", "C", "D", "E",
            "F", "G", "H", "I"
            ]

for region in regions:

    # Read the GeoJSON file into a GeoDataFrame
    gdf = gpd.read_file(f"../data/Chaos_{region}/geojson/Chaos_{region}.json",
                        driver='GeoJSON',
                        ignore_geometry=False
                        )
    # Explode MultiPolygons into individual Polygons
    gdf_exploded = gdf.explode(index_parts=False)

    if 'level_0' in gdf_exploded.columns:
        gdf_exploded.drop('level_0', axis=1, inplace=True)

    if 'level_1' in gdf_exploded.columns:
        gdf_exploded.drop('level_1', axis=1, inplace=True)

    gdf_exploded.set_crs(crs=pyproj.CRS.from_wkt('GEOGCS["GCS_Europa_2000",DATUM["D_Europa_2000",SPHEROID["Europa_2000_IAU_IAG",1562090.0,0.0]],PRIMEM["Reference_Meridian",0.0],UNIT["Degree",0.0174532925199433]]'),
                inplace=True,
                allow_override=True)

    # Drop the column you want to remove (replace 'column_to_drop' with the actual column name)
    columns_to_drop = ['Area_km', 'Area__km_', 'Lin_Feat_Amnt',
                       'Lin_Feat_A', 'Shape_Area', 'Id', 'OBJECTID',
                       'type', 'Shape_Length', 'Shape_Leng'
                       ]

    for col in columns_to_drop:
        if col in gdf_exploded.columns:
            gdf_exploded.drop(columns=[col], axis=1, inplace=True)

    # Check for rows where 'Block_Plat' is not equal to 'Plate' or 'Knob'
    filtered_gdf = gdf_exploded[(gdf_exploded['Block_Plat'] != 'Plate') & (gdf_exploded['Block_Plat'] != 'Knob')]
    if len(filtered_gdf) > 0:
        print(f"WARNING: {len(filtered_gdf)} rows where 'Block_Plat' is not equal to 'Plate' or 'Knob'")
        print(filtered_gdf)

    # Save the modified GeoDataFrame as a new GeoJSON file
    gdf_exploded.to_file(f"../data/Chaos_{region}/geojson/Chaos_{region}.json",
                         driver='GeoJSON',
                         index=True,
                         overwrite=True
                         )

    print(f"{region} saved")
```
